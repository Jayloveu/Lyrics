```python
def segment(data):
    import thulac
    cut_word = thulac.thulac(seg_only=True)
    seg_data = []
    for pair in data:
        seg_pair = []
        for string in pair:
            seg_pair.append(' '.join(list(list(zip(*cut_word.cut(string)))[0])))
        seg_data.append(seg_pair)
    return seg_data
```

```python
segment(data)
```
